```python
# returns int distance
def distance_algo(n1, n2, POD, PM, edge, agg, core, counter):
  
  distance = 0
  leftPM = 0
  rightPM = k**3/4 - 1
  leftEdge = k**3/4
  rightEdge = k**3/4 + k**2/2 - 1
  leftAgg = k**3/4 + k**2/2
  rightAgg = k**3/4 + k**2 - 1
  leftCore = k**3/4 + k**2
  rightCore = k**3/4 + k**2 + k**2/4 - 1

  # figure out the distance
#------------------------------- Case 1 (Done)----------------------------------
  # CASE 1: both PM (k=4 then 0-15)
  # (for k = 4) if n mod 15 < 16 then it is a PM
  if leftPM <= n1 <= rightPM and leftPM <= n2 <= rightPM:
    # check if same POD 
    if k == 2:
      distance = 6
      return distance
    if int(n1/(k**2/4)) == int(n2/(k**2/4)): # same POD
      # check if same edge 
      if int(n1/(k/2)) == int(n2/(k/2)): # same edge
        distance = 2
      else: # different edge
        distance = 4
    # default return 6 (different POD)
    else:
      distance = 6
    print("PM to PM")
    return distance
#------------------------------- Case 2 (Done)----------------------------------
  # CASE 2 both edge switch (k=4 then 16-23)
  elif leftEdge <= n1 <= rightEdge and leftEdge <= n2 <= rightEdge:
    # check if same POD 
    if int(n1/((k**2/4)/(k/2))) == int(n2/((k**2/4)/(k/2))): # same POD
      distance = 2
    else: # different POD
      distance = 4
    print("edge to edge")
    return distance
#------------------------------- Case 3 (Done)----------------------------------
  # CASE 3 both agg switch (k=4 then 24-31)
  elif leftAgg <= n1 <= rightAgg and leftAgg <= n2 <= rightAgg:
    # check if same POD 
    if int(n1/((k**2/4)/(k/2))) == int(n2/((k**2/4)/(k/2))): # same POD
      distance = 2
    else: # different POD
      # check if left or right agg
      #print(int((n1-leftAgg)%(k/2)), int((n2-leftAgg)%(k/2)))
      if int((n1-leftAgg)%(k/2)) == int((n2-leftAgg)%(k/2)): # if same side agg
        distance = 2
      else: # if different side agg
        distance = 4
    print("agg to agg")
    return distance
#------------------------------- Case 4 (Done)----------------------------------
  # CASE 4 both core switch (k=4 then 32-35)
  elif leftCore <= n1 <= rightCore and leftCore <= n2 <= rightCore:
    if k == 2:
      distance = 0
    if int((n1-leftCore)/((k**2/4)/(k/2))) == int((n2-leftCore)/((k**2/4)/(k/2))): # connected to same Agg? (same formula as finding same POD)
      distance = 2
    else:
      distance = 4
    print("core to core")
    return distance
#------------------------------- Case 5 ----------------------------------------
  # CASE 5 pm to edge   !!!!!!!!!!!!!!  
  elif leftPM <= n1 <= rightPM and leftEdge <= n2 <= rightEdge or leftPM <= n2 <= rightPM and leftEdge <= n1 <= rightEdge:
    # swap n1 with n2 as fomula only works one way
    if n1 > n2:
      n3 = n2
      n2 = n1
      n1 = n3
    
    print("PM to edge or edge to PM") 
    return distance
#------------------------------- Case 6 ----------------------------------------
  # CASE 6 pm to agg    !!!!!!!!!!!!!!
  elif leftPM <= n1 <= rightPM and leftAgg <= n2 <= rightAgg or leftPM <= n2 <= rightPM and leftAgg <= n1 <= rightAgg:
    # swap n1 with n2 as fomula only works one way
    if n1 > n2:
      n3 = n2
      n2 = n1
      n1 = n3
    
    print("PM to agg or agg to PM")
    return distance
#------------------------------- Case 7 (Done)----------------------------------
  # CASE 7 pm to core 
  elif leftPM <= n1 <= rightPM and leftCore <= n2 <= rightCore or leftPM <= n2 <= rightPM and leftCore <= n1 <= rightCore:
    distance = 3
    print("PM to core or core to PM")
    return distance
#------------------------------- Case 8 ----------------------------------------
  # CASE 8 edge to agg    !!!!!!!!!!!!!!
  elif leftEdge <= n1 <= rightEdge and leftAgg <= n2 <= rightAgg or leftEdge <= n2 <= rightEdge and leftAgg <= n1 <= rightAgg:
    # swap n1 with n2 as fomula only works one way
    if n1 > n2:
      n3 = n2
      n2 = n1
      n1 = n3
     
    print("edge to agg or agg to edge")
    return distance
#------------------------------- Case 9 (Done)----------------------------------
  # CASE 9 edge to core 
  elif leftEdge <= n1 <= rightEdge and leftCore <= n2 <= rightCore or leftEdge <= n2 <= rightEdge and leftCore <= n1 <= rightCore:
    distance = 2
    print("edge to core or core to edge")
    return distance
#------------------------------- Case 10 (Done)---------------------------------
  # CASE 10 agg to core 
  elif leftAgg <= n1 <= rightAgg and leftCore <= n2 <= rightCore or leftAgg <= n2 <= rightAgg and leftCore <= n1 <= rightCore:
    # swap n1 with n2 as fomula only works one way
    if n1 > n2:
      n3 = n2
      n2 = n1
      n1 = n3
    if k == 2:
      distance = 1
    # check if agg edge is same as core edge
    if int((n1-leftAgg)%(k/2)) == int((n2-leftCore)/((k**2/4)/(k/2))):
      distance = 1
    else:
      distance = 3
    print("agg to core or core to agg")
    return distance
#----------------------------- Case default ------------------------------------
  # CASE default    
  else:
    distance = -1

  return distance
```

```python
from .topology import Topology
from .util import gen_nodes, preprocess_node_positions
import numpy as np

class FatTree(Topology):

    def __init__(self, port_count, capacity_function=None):
        """
        Raises a ValueError if port_count is not even.
        :param port_count: How many ports the building switches have.
        :param capacity_function (optional, defaults to None): Function used to initialise link capacities based on their endpoints.
        """

        if port_count % 2 != 0:
            raise ValueError("There must be an even nr. of ports to construct a fat-tree")
        # Calculate the nr. of switches in each layer and store them directly in the instance of the topology
        # a spine group has as many switches as uplinks on a pod switch (i.e., ports/2)
        self.port_count = port_count
        pods = port_count
        self.tor_switches = int(pods * (port_count / 2))
        self.aggregation_switches = int(pods * (port_count / 2))
        self.core_switches = int(pow(port_count / 2, 2))
        self.tor_idx_range = range(1, self.tor_switches + 1)
        last_tor_idx = self.tor_idx_range[-1]
        self.aggregation_idx_range = range(last_tor_idx + 1, last_tor_idx + 1 + self.aggregation_switches)
        last_agg_idx = self.aggregation_idx_range[-1]
        self.core_idx_range = range(last_agg_idx + 1, last_agg_idx + 1 + self.core_switches)
        indices = [self.tor_idx_range, self.aggregation_idx_range, self.core_idx_range]
        super().__init__(indices, "FatTree_" + str(port_count), capacity_function)
        
```
