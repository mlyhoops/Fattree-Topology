```python
#User input for K variable

k1 = ""

while k1 != "exit":
  k1 = int(input('Choose an even k value\n'))
  if (k1 % 2) == 0: 
    k = k1
    # print(k)
    break
  else: 
    print ('That is not an even k value')

counter = 0
PM = 0
edge = 0
agg = 0
core = 0
POD = 0
edgeTotal = 0
aggTotal = 0

# default case when k = 2
if k == 2:
  POD = 0
  PM = 1
  edge = 2
  agg = 3
  core = 4
  counter = 4
else: # all other k values
  # print all PODs
  for i in range(int(k)):
    print("POD ID#: " + chr(80) + str(i))
    POD = i

  # print ID for all PM
  for i in range(int(k**3/4)):
    print("PM ID#: " + str(i))
    counter = i

  PM = counter

  # print ID for all Edge Switches
  for i in range(int(k**3/4), int(k**3/4) + int(k**2/2)):
    print("Edge Switch ID#: " + str(i))
    counter = i

  edge = counter

  # print ID for all Aggregate Switches
  for i in range(int(k**3/4 + k**2/2), int(k**3/4) + int(k**2)):
    print("Aggregate Switch ID#: " + str(i))
    counter = i

  agg = counter

  # print ID for all Core Switches
  for i in range(int(k**3/4) + int(k**2), int(k**3/4) + int(k**2) + int(k**2/4)):
    print("Core Switch ID#: " + str(i))
    counter = i

  core = counter

# Prints out each group of nodes
print("POD: " + str(0) + " - " + str(POD) + "\n" +
      "PM: " + str(0) + " - " + str(PM) + "\n" +
      "Edge: " + str(PM + 1) + " - " + str(edge) +  "\n" +
      "Agg: " + str(edge + 1) + " - " + str(agg) +  "\n" +
      "Core: " + str(agg + 1) + " - " + str(core))
```

```python
# returns int distance
def distance_algo(n1, n2, POD, PM, edge, agg, core, counter):
  
  distance = 0
  leftPM = 0
  rightPM = k**3/4 - 1
  leftEdge = k**3/4
  rightEdge = k**3/4 + k**2/2 - 1
  leftAgg = k**3/4 + k**2/2
  rightAgg = k**3/4 + k**2 - 1
  leftCore = k**3/4 + k**2
  rightCore = k**3/4 + k**2 + k**2/4 - 1

  if k == 2:
    leftPM = 0
    rightPM = 1
    leftEdge = 2
    rightEdge = 2
    leftAgg = 3
    rightAgg = 3
    leftCore = 4
    rightCore = 4

  # figure out the distance
#------------------------------- Case 1 (Done)----------------------------------
  # CASE 1: both PM (k=4 then 0-15)
  # (for k = 4) if n mod 15 < 16 then it is a PM
  if leftPM <= n1 <= rightPM and leftPM <= n2 <= rightPM:
    # check if same POD 
    if k == 2:
      distance = 2
      return distance
    if int(n1/(k**2/4)) == int(n2/(k**2/4)): # same POD
      # check if same edge 
      if int(n1/(k/2)) == int(n2/(k/2)): # same edge
        distance = 2
      else: # different edge
        distance = 4
    # default return 6 (different POD)
    else:
      distance = 6
    print("PM to PM")
    return distance
#------------------------------- Case 2 (Done)----------------------------------
  # CASE 2 both edge switch (k=4 then 16-23)
  elif leftEdge <= n1 <= rightEdge and leftEdge <= n2 <= rightEdge:
    # check if same POD 
    if int((n1-leftEdge)/((k**2/4)/(k/2))) == int((n2-leftEdge)/((k**2/4)/(k/2))): # same POD
      distance = 2
    else: # different POD
      distance = 4
    print("edge to edge")
    return distance
#------------------------------- Case 3 (Done)----------------------------------
  # CASE 3 both agg switch (k=4 then 24-31)
  elif leftAgg <= n1 <= rightAgg and leftAgg <= n2 <= rightAgg:
    # check if same POD 
    if int((n1-leftAgg)/((k**2/4)/(k/2))) == int((n2-leftAgg)/((k**2/4)/(k/2))): # same POD
      distance = 2
    else: # different POD
      # check if left or right agg
      #print(int((n1-leftAgg)%(k/2)), int((n2-leftAgg)%(k/2)))
      if int((n1-leftAgg)%(k/2)) == int((n2-leftAgg)%(k/2)): # if same side agg
        distance = 2
      else: # if different side agg
        distance = 4
    print("agg to agg")
    return distance
#------------------------------- Case 4 (Done)----------------------------------
  # CASE 4 both core switch (k=4 then 32-35)
  elif leftCore <= n1 <= rightCore and leftCore <= n2 <= rightCore:
    if k == 2:
      distance = 0
      return distance
    if int((n1-leftCore)/((k**2/4)/(k/2))) == int((n2-leftCore)/((k**2/4)/(k/2))): # connected to same Agg? (same formula as finding same POD)
      distance = 2
    else:
      distance = 4
    print("core to core")
    return distance
#------------------------------- Case 5 (Done)----------------------------------
  # CASE 5 pm to edge 
  elif leftPM <= n1 <= rightPM and leftEdge <= n2 <= rightEdge or leftPM <= n2 <= rightPM and leftEdge <= n1 <= rightEdge:
    # swap n1 with n2 as formula only works one way
    if n1 > n2:
      n3 = n2
      n2 = n1
      n1 = n3
    if k == 2:
      distance = 1
      return distance
    # if same POD and same Edge - 1 else 3 else 5
    if int(n1/(k**2/4)) == int((n2-leftEdge)/((k**2/4)/(k/2))): # same POD
      # check if same edge 
      if int(n1%(k/2)) == int((n2-leftEdge)%(k/2)): # same edge
        distance = 1
      else: # different edge
        distance = 3
    else:
      distance = 5
    print("PM to edge or edge to PM") 
    return distance
#------------------------------- Case 6 (Done)----------------------------------
  # CASE 6 pm to agg
  elif leftPM <= n1 <= rightPM and leftAgg <= n2 <= rightAgg or leftPM <= n2 <= rightPM and leftAgg <= n1 <= rightAgg:
    # swap n1 with n2 as formula only works one way
    if n1 > n2:
      n3 = n2
      n2 = n1
      n1 = n3
    if k == 2:
      distance = 2
      return distance
    # if same POD then 2 else 4
    if int(n1/(k**2/4)) == int((n2-leftAgg)/((k**2/4)/(k/2))):
      distance = 2
    else:
      distance = 4
    print("PM to agg or agg to PM")
    return distance
#------------------------------- Case 7 (Done)----------------------------------
  # CASE 7 pm to core 
  elif leftPM <= n1 <= rightPM and leftCore <= n2 <= rightCore or leftPM <= n2 <= rightPM and leftCore <= n1 <= rightCore:
    distance = 3
    print("PM to core or core to PM")
    return distance
#------------------------------- Case 8 (Done)----------------------------------
  # CASE 8 edge to agg
  elif leftEdge <= n1 <= rightEdge and leftAgg <= n2 <= rightAgg or leftEdge <= n2 <= rightEdge and leftAgg <= n1 <= rightAgg:
    # swap n1 with n2 as formula only works one way
    if n1 > n2:
      n3 = n2
      n2 = n1
      n1 = n3
    # if same POD then 1 else 3
    if int((n1-leftEdge)/((k**2/4)/(k/2))) == int((n2-leftAgg)/((k**2/4)/(k/2))): # same POD
      distance = 1
    else: # different POD
      distance = 3
    print("edge to agg or agg to edge")
    return distance
#------------------------------- Case 9 (Done)----------------------------------
  # CASE 9 edge to core 
  elif leftEdge <= n1 <= rightEdge and leftCore <= n2 <= rightCore or leftEdge <= n2 <= rightEdge and leftCore <= n1 <= rightCore:
    distance = 2
    print("edge to core or core to edge")
    return distance
#------------------------------- Case 10 (Done)---------------------------------
  # CASE 10 agg to core 
  elif leftAgg <= n1 <= rightAgg and leftCore <= n2 <= rightCore or leftAgg <= n2 <= rightAgg and leftCore <= n1 <= rightCore:
    # swap n1 with n2 as formula only works one way
    if n1 > n2:
      n3 = n2
      n2 = n1
      n1 = n3
    if k == 2:
      distance = 1
      return distance
    # check if agg edge is same as core edge
    if int((n1-leftAgg)%(k/2)) == int((n2-leftCore)/((k**2/4)/(k/2))):
      distance = 1
    else:
      distance = 3
    print("agg to core or core to agg")
    return distance
#----------------------------- Case default ------------------------------------
  # CASE default    
  else:
    distance = -1

  return distance
```
